# Statistical Business Analysis (Week 7)

## Load Data

```
import pandas as pd
sales = pd.read_csv('/mnt/data/sales_data.csv')
churn = pd.read_csv('/mnt/data/customer_churn.csv')
sales.head()
```

## Descriptive Statistics

```
sales.describe()
```

## Distribution

```
import matplotlib.pyplot as plt
sales.select_dtypes('number').hist(figsize=(8,6)); plt.show()
```

## Correlation

```
sales.select_dtypes('number').corr()
```

## Hypothesis Testing

```
from scipy import stats
stats.ttest_1samp(sales.select_dtypes('number').iloc[:,0], sales.select_dtypes('number').iloc[:,0].mean())
```

## Confidence Interval

```
from scipy import stats
col=sales.select_dtypes('number').iloc[:,0]
mean=col.mean(); sem=stats.sem(col)
stats.t.interval(0.95, len(col)-1, loc=mean, scale=sem)
```

## Regression

```
import statsmodels.api as sm
X=sales.select_dtypes('number').iloc[:,1:]
y=sales.select_dtypes('number').iloc[:,0]
Xc=sm.add_constant(X)
sm.OLS(y, Xc, missing='drop').fit().summary()
```
